_Utilizando a base "realtor-data.csv"_

Gerar uma amostra com os seguintes requisitos:
- amostra aleatória com 1.000 registros, sendo:
    - 250 do estado de Massachusetts
    - 250 do estado de Puerto Rico
    - 250 do estado de Connecticut
    - 250 dos estados Rhode Island e New York
- somente casas for_sale
- sem duplicidades de status, city, state e price
- na remoção de duplicados priorizar as casas maiores.
- Valide a amostra criada.

```python
df_realtors = pd.read_csv('../datasets/realtor-data.csv')

df_realtors
```

```python
df_realtors.info()
df_realtors.shape
```

```python
df_realtors_filtered = df_realtors[df_realtors['status'] == 'for_sale']

df_realtors_filtered
```

```python
df_realtors_filtered.groupby('status').count()[['price']]
```

```python
df_realtors_filtered.groupby('state').count()[['price']].sort_values(by='price', ascending=False)
```

```python
df_realtors_filtered = df_realtors_filtered.sort_values('house_size')
df_realtors_filtered.head()
```

```python
df_realtors_deduplicated = df_realtors_filtered.drop_duplicates(['status','city','state','price'],keep='last')

df_realtors_deduplicated.info()
```

```python
df_per_state = df_realtors_deduplicated[df_realtors_deduplicated['state'].isin(['Massachusetts','Connecticut','Rhode Island','New York'])].sort_values('state')

df_per_state
```

```python
massachusetts = df_realtors_deduplicated[df_realtors_deduplicated['state']=="Massachusetts"]
massachusetts
```

```python
connecticut = df_realtors_deduplicated[df_realtors_deduplicated['state']=="Connecticut"]
connecticut
```

```python
rhode_island = df_realtors_deduplicated[df_realtors_deduplicated['state']=="Rhode Island"]
rhode_island
```

```python
new_york = df_realtors_deduplicated[df_realtors_deduplicated['state']=="New York"]
new_york
```

```python
exercicio = pd.concat([
    massachusetts.sample(n=250,random_state=43),
    connecticut.sample(n=250,random_state=43),
    rhode_island.sample(n=250,random_state=43),
    new_york.sample(n=250,random_state=43)
])

exercicio.to_csv('resposta.csv')

```
