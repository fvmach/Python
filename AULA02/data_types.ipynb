# Python Data Types Cheatsheet

## Variables

```python
#Python  Variables

var_int = 10
var_float = 0.45
var_string = 'ahoy!'
var_bool = True
```

```python
var_array = [var_int, var_float, var_string, var_bool]
for x in var_array:
    print(type(x))
```

## Advanced Data Types

### Lists

```python
#Lists (mutable arrays)
another_list = [1,2,3,4,5]
my_list =[42, 'Om', True, another_list]

print(my_list)

my_list[1] = input('Quem vem lá?')

print('Troque aqui hein?', my_list, 'Deu pra ver?')

print(type(my_list[1]))

my_list.append('este caba é novo, visse?') #adds to next to last index

print(my_list)

my_list.pop() #removes the last item
print(my_list)

my_list.pop(1) #indexable
print(my_list)

sorted_list = sorted(my_list) #sort() and sorted() only work with comparable elements
print(my_list)
```

### Tuples

```python
#Tuples (immutable arrays)
my_tuple = (42, 'Om', True, my_list)

print(my_tuple)

my_tuple(1) = input('Vai faiá?')

print(my_tuple)
```

### Dictionaries

```python
#Dictionaries (Key-value pairs. Keys are immutable and unique. Values are not.)

my_dict = {
    "id":"00001", 
    "name":"Mariana Mendes", 
    "active_customer":True,
    "age": 26,
    "CLV": 678980.34,
    "Currency": "USD"  
}

name = my_dict['name'].split()
first_name = name[0]
last_name = name[1]

age = my_dict['age']

print(f"Thanks for subscribing, {first_name}!")
print(f"Hey there, {first_name}! Happy birthday! {age} years old is your lucky year =)")
```

### Sets

```python
#Sets (unordered lists of unique items)

my_set = {'00001','00002','00003'}
print(my_set)

my_automated_set = {
    str(i).zfill(5) for i in range(1,100000)
}
print(my_automated_set)
```

### Bytes

```python
#Bytes (Immutable sequences of bytes (for bin data))

my_bytes = b'Fernando Vieira Machado'
print(f"My binary name is, {my_bytes}")
```

### Byte Arrays

```python
#Same as Bytes, but mutable
my_byte_array = bytearray(b'Hello')
```
